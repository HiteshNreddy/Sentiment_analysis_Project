```python
# ============================================================
# NOTEBOOK CELL 4: Word count comparison
# ============================================================

df_processed["original_words"] = df_processed["review"].apply(
    lambda x: len(str(x).split())
)

df_processed["cleaned_words"] = df_processed["clean_review"].apply(
    lambda x: len(str(x).split())
)

print("Average BEFORE cleaning:",
      int(df_processed["original_words"].mean()))

print("Average AFTER cleaning:",
      int(df_processed["cleaned_words"].mean()))

reduction = (
    1 - df_processed["cleaned_words"].mean()
    / df_processed["original_words"].mean()
) * 100

print("Word reduction:", int(reduction), "%")

df_processed.drop(columns=["original_words", "cleaned_words"], inplace=True)
```

```python
import sys
import os

sys.path.append(os.path.abspath(".."))

import pandas as pd
import numpy as np

from src.features import (
    build_tfidf_vectorizer,
    split_data,
    fit_transform_tfidf,
    save_vectorizer,
    inspect_vocabulary
)

df = pd.read_csv('../data/processed/cleaned_reviews.csv')

print(f"Loaded cleaned dataset: {df.shape[0]} rows\n")

print("Splitting data into train and test")

X_train, X_test, y_train, y_test = split_data(
    df,
    text_column='clean_review',
    label_column='sentiment',
    test_size=0.2
)

print("Training samples:", len(X_train))
print("Test samples:", len(X_test))

vectorizer = build_tfidf_vectorizer()

X_train_tfidf, X_test_tfidf, vectorizer = fit_transform_tfidf(
    vectorizer,
    X_train,
    X_test
)

print("TF-IDF train matrix:", X_train_tfidf.shape)
print("TF-IDF test matrix:", X_test_tfidf.shape)

save_vectorizer(vectorizer)

inspect_vocabulary(vectorizer)
```

```python
# ============================================================
# NOTEBOOK CELL: Build and Fit TF-IDF
# ============================================================

# Build TF-IDF vectorizer
vectorizer = build_tfidf_vectorizer(
    max_features=10000,
    ngram_range=(1, 2),
    min_df=2,
    max_df=0.95,
    sublinear_tf=True
)

# Fit vectorizer on training data and transform both sets
X_train_tfidf, X_test_tfidf, vectorizer = fit_transform_tfidf(
    vectorizer,
    X_train,
    X_test
)

# -----------------------------------------------------------
# Inspect matrices
# -----------------------------------------------------------

print("\n" + "=" * 55)
print("TF-IDF MATRIX SUMMARY")
print("=" * 55)

print("\nTraining matrix shape :", X_train_tfidf.shape)
print("  Rows    :", X_train_tfidf.shape[0])
print("  Columns :", X_train_tfidf.shape[1])

print("\nTest matrix shape :", X_test_tfidf.shape)
print("  Rows    :", X_test_tfidf.shape[0])
print("  Columns :", X_test_tfidf.shape[1])

# -----------------------------------------------------------
# Sparsity calculation
# -----------------------------------------------------------

total_cells = X_train_tfidf.shape[0] * X_train_tfidf.shape[1]
nonzero_cells = X_train_tfidf.nnz

sparsity = (1 - nonzero_cells / total_cells) * 100

print("\nMatrix sparsity")
print("  Total cells    :", total_cells)
print("  Non-zero cells :", nonzero_cells)
print("  Sparsity       :", round(sparsity, 2), "%")

# -----------------------------------------------------------
# Memory usage
# -----------------------------------------------------------

sparse_size = X_train_tfidf.data.nbytes / (1024 * 1024)

print("\nMemory usage (sparse format):", round(sparse_size, 2), "MB")

dense_size = total_cells * 8 / (1024 * 1024)

print("Memory if dense:", round(dense_size, 2), "MB")

saving = (1 - sparse_size / dense_size) * 100

print("Memory saved using sparse matrix:", round(saving, 2), "%")
```

```python
# ============================================================
# NOTEBOOK CELL: Save the vectorizer and dataset splits
# ============================================================

import os
import numpy as np

# Create required folders if they don't exist
os.makedirs("models", exist_ok=True)
os.makedirs("data/processed", exist_ok=True)

# Save the trained TF-IDF vectorizer
save_vectorizer(vectorizer, path="models/tfidf_vectorizer.pkl")

# Save train/test splits for later training step
np.save("data/processed/X_train.npy", X_train)
np.save("data/processed/X_test.npy", X_test)

np.save("data/processed/y_train.npy", y_train)
np.save("data/processed/y_test.npy", y_test)

print("\nAll feature extraction outputs saved:")
print("models/tfidf_vectorizer.pkl  ← fitted TF-IDF vectorizer")
print("data/processed/X_train.npy   ← training text")
print("data/processed/X_test.npy    ← test text")
print("data/processed/y_train.npy   ← training labels")
print("data/processed/y_test.npy    ← test labels")
```

```python
# ============================================================
# NOTEBOOK CELL: Visualize TF-IDF for one review
# ============================================================

print("TF-IDF transformation example on a single review:")
print("=" * 60)

# Pick one review from training set
sample_review = X_train[0]
sample_label  = y_train[0]
sample_tfidf  = X_train_tfidf[0]   # TF-IDF vector row

print("\nOriginal clean review :", sample_review[:80] + "...")
print("Sentiment label       :", sample_label)

# Get feature names
feature_names = vectorizer.get_feature_names_out()

# Non-zero indices and scores
nonzero_indices = sample_tfidf.nonzero()[1]
nonzero_scores  = sample_tfidf.data

# Sort by score (highest first) and take top 15
sorted_idx = np.argsort(nonzero_scores)[::-1][:15]

print("\nTop 15 TF-IDF features for this review:")
print(f"{'Word/Phrase':<28} {'TF-IDF Score'}")
print("-" * 42)

for idx in sorted_idx:
    col   = nonzero_indices[idx]
    score = nonzero_scores[idx]
    print(f"{feature_names[col]:<26} {score:.4f}")

print("\nTotal non-zero features in this review :", len(nonzero_indices))
print("Total vocabulary size                  :", len(feature_names))
print("Features with score = 0 in this review :", len(feature_names) - len(nonzero_indices))
```

```python
# ============================================================
# NOTEBOOK CELL: Load data and train all three models
# ============================================================

import sys
import os

# Add project root to Python path (fixes "No module named src")
sys.path.append(os.path.abspath(".."))

import numpy as np
import joblib

from src.train import train_all_models

# ------------------------------------------------------------
# Load saved train/test data
# ------------------------------------------------------------
print("Loading training data...")

X_train = np.load("../data/processed/X_train.npy", allow_pickle=True)
X_test  = np.load("../data/processed/X_test.npy", allow_pickle=True)

y_train = np.load("../data/processed/y_train.npy", allow_pickle=True)
y_test  = np.load("../data/processed/y_test.npy", allow_pickle=True)

print(f"X_train: {len(X_train)} reviews")
print(f"X_test : {len(X_test)} reviews")
print(f"y_train: {len(y_train)} labels")

# ------------------------------------------------------------
# Load the fitted TF-IDF vectorizer
# ------------------------------------------------------------
vectorizer = joblib.load("../models/tfidf_vectorizer.pkl")

# Transform text into TF-IDF matrices
X_train_tfidf = vectorizer.transform(X_train)
X_test_tfidf  = vectorizer.transform(X_test)

print("\nTF-IDF matrix loaded:")
print(f"  X_train_tfidf shape: {X_train_tfidf.shape}")
print(f"  X_test_tfidf shape : {X_test_tfidf.shape}")

# ------------------------------------------------------------
# Train all models (Naive Bayes, Logistic Regression, SVM)
# ------------------------------------------------------------
models = train_all_models(X_train_tfidf, y_train)

print("\nTraining finished.")
print("Saved models:")
print("  ../models/naive_bayes_model.pkl")
print("  ../models/logistic_regression_model.pkl")
print("  ../models/svm_model.pkl")
```
